 * **Проверка гипотез**

Из данных мы имеем следующее:

 * **name** - название фильма.
 * **rating** - возврастной рейтинг (R, PG, PG-13 и тд).
 * **genre** - жанр фильма.
 * **year** - год выпуска фильма.
 * **released** - дата проката фильма, представлена в виде строки(June 13, 1980 (United States)).

 * **director** - режиссёр.

 * **country** - страна производства.
 * **budget** - бюджет фильма, у некоторых его нет, соответственно, бюджет такого фильма равен нулю.
 * **gross** - сборы с показа.
 * **company** - компания, которая производила съёмки фильма.
 * **runtime** - продолжительность фильма в минутах.

### Топ-5 жанров по кол-ву снятых фильмов.

```python
# Возвращает кол-во записей для каждого жанра
def absolute_value(val):
    genre_size = np.array(top_genres)
    result  = np.round(val/100.*genre_size.sum(), 0)
    return result

# Берём топ-5 жанров из нашего датасета. 
# В итоге должна получиться серия где в качестве ключа жанр, а в качстве значения кол-во фильмов по этому жанру.
top_genres = formated_movies['genre'].value_counts().head(5)

# Формируем фигуру, в которой будут храниться диаграмы.
figure, (percent_pie, value_pie) = plt.subplots(1,2,figsize=(15,8))
figure.suptitle('Топ 5 жанров по кол-ву снятых фильмов', fontsize=24)

# Диаграма с процентами
percent_pie.pie(top_genres, labels = top_genres.keys(), autopct='%1.1f%%')
percent_pie.set_title('Проценты',fontsize = 18)

# Диаграма со значениями
value_pie.pie(top_genres, labels = top_genres.keys(), autopct=absolute_value)
value_pie.set_title('Значение', fontsize = 18)

plt.show()
```

Как мы видим, больше всего снимали следующие жанры:
  * 1. Комедия
  * 2. Экшен
  * 3. Драма
  * 4. Криминал
  * 5. Приключения

Давайте изобразим на диаграме кол-во снятых фильмов и доход жанра по годам.

```python
# Таблица с количеством за год по каждому жанру.
film_count_by_year = formated_movies.groupby(['year', 'genre']).agg('count').reset_index()
film_count_by_year.drop(film_count_by_year.columns[3:], axis = 1, inplace=True)
film_count_by_year.rename({'name':'count'},axis = 1, inplace=True)

# Таблица со средним доходом за год по каждому жанру
film_mean_gross_by_year = formated_movies.groupby(['year', 'genre']).agg('mean').reset_index()
film_mean_gross_by_year.drop(['score', 'votes(thousand)', 'runtime'], axis=1, inplace=True)
film_mean_gross_by_year.rename({'gross(millions)': 'mean_gross'}, axis=1, inplace=True)

fig, axes = plt.subplots(5, 2, figsize=(16,32))

# Диаграммы комедии.
comedy_count_by_year = film_count_by_year.query("genre == 'Comedy'")
axes[0, 0].bar(comedy_count_by_year['year'], comedy_count_by_year['count'])
axes[0,0].set_title('Комедия', fontsize=16)
axes[0, 0].set_xlabel('Год')
axes[0, 0].set_ylabel('Кол-во снятых фильмов')

comedy_mean_by_year = film_mean_gross_by_year.query("genre == 'Comedy'")
axes[0, 1].bar(comedy_mean_by_year['year'], comedy_mean_by_year['mean_gross'])
axes[0, 1].set_title('Комедия', fontsize=16)
axes[0, 1].set_xlabel('Год')
axes[0, 1].set_ylabel('Средняя прибыль')

# Диаграммы экшена.
action_count_by_year = film_count_by_year.query("genre == 'Action'")
axes[1, 0].bar(action_count_by_year['year'], action_count_by_year['count'])
axes[1, 0].set_title('Экшен', fontsize=16)
axes[1, 0].set_xlabel('Год')
axes[1, 0].set_ylabel('Кол-во снятых фильмов')

action_mean_by_year = film_mean_gross_by_year.query("genre == 'Action'")
axes[1, 1].bar(action_mean_by_year['year'], action_mean_by_year['mean_gross'])
axes[1, 1].set_title('Экшен', fontsize=16)
axes[1, 1].set_xlabel('Год')
axes[1, 1].set_ylabel('Средняя прибыль')

# Диаграммы драмы.
drama_count_by_year = film_count_by_year.query("genre == 'Drama'")
axes[2, 0].bar(drama_count_by_year['year'], drama_count_by_year['count'])
axes[2, 0].set_title('Драма', fontsize=16)
axes[2, 0].set_xlabel('Год')
axes[2, 0].set_ylabel('Кол-во снятых фильмов')

drama_mean_by_year = film_mean_gross_by_year.query("genre == 'Drama'")
axes[2, 1].bar(drama_mean_by_year['year'], drama_mean_by_year['mean_gross'])
axes[2, 1].set_title('Драма', fontsize=16)
axes[2, 1].set_xlabel('Год')
axes[2, 1].set_ylabel('Средняя прибыль')

# Диаграммы криминала.
crime_count_by_year = film_count_by_year.query("genre == 'Crime'")
axes[3, 0].bar(crime_count_by_year['year'], crime_count_by_year['count'])
axes[3, 0].set_title('Криминал', fontsize=16)
axes[3, 0].set_xlabel('Год')
axes[3, 0].set_ylabel('Кол-во снятых фильмов')

crime_mean_by_year = film_mean_gross_by_year.query("genre == 'Crime'")
axes[3, 1].bar(crime_mean_by_year['year'], crime_mean_by_year['mean_gross'])
axes[3, 1].set_title('Криминал', fontsize=16)
axes[3, 1].set_xlabel('Год')
axes[3, 1].set_ylabel('Средняя прибыль')

# Диаграммы приключения.
adventure_count_by_year = film_count_by_year.query("genre == 'Adventure'")
axes[4, 0].bar(adventure_count_by_year['year'], adventure_count_by_year['count'])
axes[4, 0].set_title('Приключение', fontsize=16)
axes[4, 0].set_xlabel('Год')
axes[4, 0].set_ylabel('Кол-во снятых фильмов')

adventure_mean_by_year = film_mean_gross_by_year.query("genre == 'Adventure'")
axes[4, 1].bar(adventure_mean_by_year['year'], adventure_mean_by_year['mean_gross'])
axes[4, 1].set_title('Приключения', fontsize=16)
axes[4, 1].set_xlabel('Год')
axes[4, 1].set_ylabel('Средняя прибыль')

plt.show()
```

 * **Комедия**
   * Распределение среди кол-ва снятых фильмов кое-как стремиться к **нормальному**.

### Любимый жанр исходя из оценок зрителей по годам.

```python
favorite_genres_by_year = formated_movies.groupby(['genre','year']) \
                                        .agg('mean').sort_values(['year', 'votes(thousand)']).reset_index()
favorite_genres_by_year.drop(['score', 'gross(millions)', 'runtime'], axis=1, inplace=True)
max_votes_by_year = favorite_genres_by_year.groupby(['year']).agg('max').reset_index()
favorite_genres_by_year = favorite_genres_by_year.merge(max_votes_by_year, how='inner', on=['year','votes(thousand)']) \
                        .drop(['genre_y'], axis=1) \
                       
favorite_genres_by_year.rename({'genre_x':'best_genre', 'year_x':'year'}, axis=1, inplace=True)
favorite_genres_by_year.drop_duplicates(inplace=True)
favorite_genres_by_year = favorite_genres_by_year.sort_values('year')
favorite_genres_by_year
```

```python
plt.plot(favorite_genres_by_year['best_genre'].value_counts())
plt.title("График количества номинаций лучших жанров по оценкам зрителей", fontsize=18)
plt.xlabel("Жанры", fontsize=16)
plt.ylabel("Кол-во номинаций", fontsize=16)
plt.show()
```

```python
def show_graphics() :
    genres = favorite_genres_by_year['best_genre'].unique()
    for i in genres:
        genre_years = favorite_genres_by_year['year'][favorite_genres_by_year['best_genre'] == i]
        genre_gross = favorite_genres_by_year['votes(thousand)'][favorite_genres_by_year['best_genre'] == i]
        plt.scatter(genre_years, genre_gross)
        
    plt.legend(genres)
    plt.show()
    
show_graphics()
```

### Лучший жанр в году по мнению критиков.

```python
best_genre_by_score = formated_movies.groupby(['genre', 'year']) \
                                      .agg('mean').sort_values(['year', 'score']).reset_index()
best_genre_by_score.drop(['votes(thousand)', 'gross(millions)', 'runtime'], axis=1, inplace=True)

max_score_genre = best_genre_by_score.groupby('year').agg('max').reset_index()
best_genre_by_score = best_genre_by_score.merge(max_score_genre, how='inner', on=['score', 'year']) \
                                            .drop(['genre_y'], axis=1)
                                          
best_genre_by_score.drop_duplicates(inplace=True)
best_genre_by_score.rename({'genre_x':'best_genre'}, axis=1, inplace=True)
best_genre_by_score
```

```python
plt.plot(best_genre_by_score['best_genre'].value_counts())
plt.title("График количества номинаций лучших жанров по оценкам критиков", fontsize=18)
plt.xlabel("Жанры", fontsize=16)
plt.ylabel("Кол-во номинаций", fontsize=16)
plt.show()
```

### Средняя прибыль за картину по возврастному рейтингу.

```python
mean_gross_by_rating = formated_movies.groupby('rating').agg('mean').reset_index()
mean_gross_by_rating.drop(['year', 'score', 'votes(thousand)', 'runtime'], axis=1, inplace=True)
mean_gross_by_rating.sort_values('gross(millions)')
```

```python
sum_gross_by_rating = formated_movies.groupby('rating').agg('sum').reset_index()
sum_gross_by_rating.drop(['year', 'score', 'votes(thousand)', 'runtime'], axis=1, inplace=True)
sum_gross_by_rating.sort_values('gross(millions)')
```

Мы видим, что чем безобиднее контент в фильме, тем больше у него прибыль.
   * **G** - Фильм демонстрируется без ограничений. Данный рейтинг показывает, что оценённый фильм не содержит ничего, что большинство родителей могло бы посчитать неприемлемым для просмотра или прослушивания даже самыми маленькими детьми. Обнажение, сексуальные сцены и сцены приёма наркотиков отсутствуют; насилие минимально; могут употребляться выражения, выходящие за пределы вежливой беседы, но только те, которые постоянно встречаются в повседневной речи. Более грубая лексика в фильмах с рейтингом G употребляться не может.

## Проверка гипотез

Для проверки гипотез я буду использовать модуль **stats** в библиотеке **SciPy** 

```python
from scipy import stats
```

### 1.  Правда ли что жанр Ужасы более прибыльнее, чем жанр Комедия?

Составим варианты гипотез:
 * **H0** Картина жанра Ужасы по прибыле такие же как картины жанра Комедия.
 * **Н1** Картина жанра Ужасы более прибыльная. 

Здесь я буду сравнивать по средним значениям двух жанров. Для этого хорошо подойдёт тест Стьдента, его мы будем использовать. Важно, чтобы две наши группы имели одинаковое кол-во наблюдений, поэтому сделаем выборку из 30 наблюдений в каждой.

```python
def check_p_value(p_value):
    print(f'p = {p_value}')
    
    if p_value > 0.05:
        print('Принимаем нулевую гипотезу\n')
    else:
        print('Отклоняем нулевую гипотезу\n')
        

# Возьмём сначала все фильмы по жанрам.
comedy_movies = formated_movies.query("genre == 'Comedy'")
horror_movies = formated_movies.query("genre == 'Horror'")

# Делаем случайную выборку из 30 фильмов.
sample_comedy = comedy_movies.sample(30)
sample_horror = horror_movies.sample(30)
    
stat, p_value = stats.ttest_ind(sample_horror['gross(millions)'], sample_comedy['gross(millions)'])
check_p_value(p_value)
```

### 2. Действительно ли G фильмы более прибыльнее, чем PG-13 фильмы?

Составим варианты гипотез:
 * **H0** G фильмы прибыле такие же как картины PG-13.
 * **Н1** Картина жанра G более прибыльная. 

```python
G_movies = formated_movies.query("rating == 'G'")
PG_13_movies = formated_movies.query("rating == 'PG-13'")


# Делаем случайную выборку из 50 фильмов.
sample_g_movies = G_movies.sample(50)
sample_pg_13_movies = PG_13_movies.sample(50)
    
stat, p_value = stats.ttest_ind(sample_g_movies['gross(millions)'], sample_pg_13_movies['gross(millions)'])
check_p_value(p_value)
```

### 3. Картины Квентина Тарантино критикам нравятся больше чем картины Стивена Спилберга.

Составим варианты гипотез:
 * **H0** Картины Квентина Тарантино критики оценивают так же как картины Стивена Спилберга.
 * **Н1** Картины Квентина Тарантино критикам нравятся больше чем картины Стивена Спилберга.

```python
tarantino_movies = formated_movies.query("director == 'Quentin Tarantino'")
spielberg_movies = formated_movies.query("director == 'Steven Spielberg'")

# Делаем случайную выборку из 30 фильмов.
sample_tarantino_movies = tarantino_movies.sample(6)
sample_spielberg_movies = spielberg_movies.sample(6)
    
stat, p_value = stats.ttest_ind(sample_tarantino_movies['score'], sample_spielberg_movies['score'])
check_p_value(p_value)
```

### Резюмирование проверки гипотез.

 * Фильмы ужасов не отличаются по прибыли от комедийных фильмов.
 * Фильмы с возврастным рейтингом G одинаковы по прибыли с фильмами рейтингом PG-13.
 * Картины Квентина Тарантино критикам нравятся больше чем картины Стивена Спилберга.
